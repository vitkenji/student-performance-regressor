```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('features.csv')
```

```python
print(df.columns)
```

```python
print(f"Média da idade:{df['age'].mean()}")
print(f"Porcentagem de endereços urbanos: {(df['urban_address'] == 1).sum() / len(df) * 100:.2f}")
print(f"Porcentagem de homens: {(df['is_male'] == 1).sum() / len(df) * 100:.2f}")
```

```python
sns.scatterplot(data=df, x='age', y='target')
plt.xlabel('age')
plt.ylabel('G3')
plt.show()
```

```python
sns.scatterplot(data=df, x='Medu', y='target')
plt.xlabel('mother education')
plt.ylabel('G3')
plt.show()
```

```python
sns.scatterplot(data=df, x='Fedu', y='target')
plt.xlabel('father education')
plt.ylabel('G3')
plt.show()
```

```python
sns.countplot(data=df, x='studytime')
plt.xlabel('Tempo de Estudo Semanal (nível)')
plt.ylabel('Número de Alunos')
plt.title('Distribuição do Tempo de Estudo')
plt.xticks([0, 1, 2, 3], ['<2h', '2-5h', '5-10h', '>10h']) 
plt.show()
```

```python
sns.violinplot(data=df, x='studytime', y='target', inner='quartile')
plt.xlabel('Tempo de Estudo Semanal (nível)')
plt.ylabel('Nota Final (G3)')
plt.title('Distribuição das Notas por Tempo de Estudo')
plt.xticks([0, 1, 2, 3], ['<2h', '2-5h', '5-10h', '>10h'])
plt.show()
```

```python
sns.scatterplot(data=df, x='absences', y='target')
plt.xlabel('absences')
plt.ylabel('G3')
plt.show()
```
